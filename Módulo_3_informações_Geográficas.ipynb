# Mapeando Dados com Folium

O pacote `folium` é um dos mais utilizados para plotar informações geográficas.

Para criar um gráafico, basta instanciar um objeto do tipo `Map` e informar as coordenadas iniciais e o nível de zoom.

```
# importar os pacotes necessários
import pandas as pd
import folium
```

```
# A bilioteca permite abrir um mapa, passando a localização central e o zoom inicial
# Se abiri o google maps, o final do link é a localização e o zoom no qual a página esta (https://www.google.com.br/maps/@-21.3753915,-45.5215608,14.25z)
mapa = folium.Map(location=[-21.3753915,-45.5215608],
                  zoom_start=14,)

mapa
```

Para mostrar como plotar as informações de latitude e longitude contidas em um *DataFrame*, vou usar as informações da prefeitura de Recife com as Estações de Bikes disponíveis na cidade.

Os dados originais podem ser obtidos diretamente [clicando neste link](http://dados.recife.pe.gov.br/dataset/ciclovias-ciclofaixas-estacoes-de-aluguel-de-bikes-e-rotas/resource/e6e4ac72-ff15-4c5a-b149-a1943386c031).

```
# O dataframe esta delimitado por ; e não por , -> Necessário informar o delimitador (delimiter=';')
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/curso_data_science_na_pratica/master/modulo_03/estacoesbike.csv",
                 delimiter=';')
df.head()
```

```
# extraindo os valores do dataframe para variáveis -> Criação de variáveis
latitude = df.latitude.values
longitude = df.longitude.values
nomes = df.nome.values
```

```
# criando um mapa
mapa = folium.Map(location=[-8.0298747,-34.9670771],
                  zoom_start=11, )

# adicionando pontos (markers) ao mapa
# É necessário informar quais variaveis são latitude, longitude e nome do ponto, por isso instanciamos as variáveis antes
# É necessário também informar o tipo dos dados de latitude e longitude
# 'popup=nome' foi criado para apresentar o nome do ponto quando clica-se no marcador
for lat, lon, nome in zip(latitude, longitude, nomes):
    folium.Marker(location=[float(lat), float(lon)], popup=nome).add_to(mapa)

# exibir mapa final com os pontos
mapa
```
